```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Lendo o arquivo `raw.csv`

```python
pathraw = "raw/heart.csv"
pathprocessed = {"age":  "processed/heart-missing-age.csv",
                 "chol": "processed/heart-missing-chol.csv",
                 "sex":  "processed/heart-missing-sex.csv"}
```

```python
dt_raw = pd.read_csv(pathraw)
dt_raw.describe()
```

### Lendo o arquivo `heart-missing-age.csv`

```python
dt_age = pd.read_csv(pathprocessed["age"])
dt_age.describe()
```

```python
print("Dados de age")
plt.figure(figsize=(16,6))
dt_raw.age.sort_values().plot.hist(label="raw")
dt_age.age.sort_values().plot.hist(label="processed")
plt.legend()
plt.show()
```

Analisando estatisticas da coluna age, aparentemente não houve uma retirada para valor especifico de age. O histograma mostra que os valores retirados eram das bins de maior frequência.

### Analisando dados faltantes de `heart-missing-age.csv`

```python
dt_age[dt_age["age"].isna()].describe()
```

```python
dt_age[dt_age["age"].isna()]
```

Chama-se bastante atenção que os dados retirados todos são para `sex=0`.
Com essa observação é possível se dizer que o processamento causou MNAR (Missing not at Random).
